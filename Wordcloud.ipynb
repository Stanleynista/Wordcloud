```
# importar os pacotes necessários
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# importar o arquivo csv em um df
df_fei = pd.read_csv("/content/drive/MyDrive/Palavras-chave-FEI.csv")
df_fiap =pd.read_csv("/content/drive/MyDrive/Palavras-chave-FIAP.csv")

# eliminar as colunas com valores ausentes
keywords_fei = df_fei.dropna(subset=['Palavras-chave'], axis=0)['Palavras-chave']
keywords_fiap = df_fiap.dropna(subset=['Palavras-chave'], axis=0)['Palavras-chave']
```

```
# exemplos de descrições para palavras-chave
display(keywords_fei.iloc[100])
display(keywords_fiap.iloc[100])
```

```
# concatenar as palavras
all_summary_fei = " ".join(s for s in keywords_fei)
all_summary_fiap = " ".join(s for s in keywords_fiap)

# ver quantidade de palavras
print("Quantidade de Palavras: {}".format(len(all_summary_fei)))
print("Quantidade de Palavras: {}".format(len(all_summary_fiap)))
```

```
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "você", "de", "ao", "os"])

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white",
                      width=1600, height=800).generate(all_summary_fei)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("fei_keywords_wordcloud.png");
```
